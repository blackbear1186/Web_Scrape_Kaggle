```python
import zipfile
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait as wait
options = webdriver.ChromeOptions()
prefs = {"download.default_directory" : "/Users/blackbear/Python_project/Selenium"}
options.add_experimental_option('prefs',prefs)



browser = webdriver.Chrome(options=options)

#Open browser and go to Kaggle.com
browser.get('https://www.kaggle.com/')

#Click Sign in button
clickSignIn = browser.find_element('link text','Sign In')
clickSignIn.click()

browser.implicitly_wait(5)
#Click login in by email link
clickEmailLogin = browser.find_element(By.XPATH, "//a[@href='/account/login?phase=emailSignIn&returnUrl=%2F']")
clickEmailLogin.click()

browser.implicitly_wait(5)
#Enter email information
emailInput = browser.find_element(By.NAME, 'email' )
emailInput.send_keys('berlin.j.1186@gmail.com')

#Enter password information
passwordInput = browser.find_element(By.NAME, 'password')
passwordInput.send_keys('password')

#After email and password is enter click submit
submitButton = browser.find_element(By.XPATH, "//button[@type='submit']")
submitButton.click()

browser.implicitly_wait(5)
#Click on datasets link
browser.find_element(By.XPATH, "//a[@href='/datasets']").click()

first_trending = browser.find_element(By.XPATH, '//div[@class="sc-fPGsrK sc-cxESYj imePyQ bFQPrT"]/a')
browser.execute_script('arguments[0].click();', first_trending) 



#Click on the download button
#Download dataset to selenium folder
download_data = browser.find_element(By.PARTIAL_LINK_TEXT,'Download')
download_data.click()

#Use sleep method to wait 10 seconds before attempting to unzip file
time.sleep(10)
#Unzip folder and extract all contents
archive = zipfile.ZipFile("archive.zip", mode='r') 
archive.extractall('/Users/blackbear/Python_project/Selenium/')    

#Create empty list
files = []

#Get zip file names and add to files list
for filename in archive.namelist():
    files.append(filename)
    
    
#Create empty list to store list of dataframes
df_list = []

#Iterate through files by index using range(len()) and add each dataframe to df_list 
for i in range(len(files)):
    if 'csv' in files[i]:
        temp_df = pd.read_csv(files[i])
        df_list.append(temp_df)

df1 = pd.DataFrame(df_list[0])

df1.head(5)

```
